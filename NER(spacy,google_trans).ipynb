```python
in short :
NER helps a machine understand "who","what",and "where" from text

example:
text:
"Rohit sharma lives in mumbai"

NER output:
Rohit sharma _> PERSON
Mumbai_> LOCATION(GPE)

why do we use NER?
we use NER to convert unstructured text into meaningful, structured data.
main reasons:
Automatically extract important information
reduce manual work in reading documents
help machines understand context,not just keywords
make text usable for analytics,ML models,and databases
```

```python
import spacy
nlp=spacy.load("en_core_web_sm")
```

```python
text="Barack Obama was born in Hawai and become President of the United States"

doc=nlp(text)

for ent in doc.ents:
    print(ent.text,"->",ent.label_)
```

```python
from googletrans import Translator

translator = Translator()

hinglish_text = "Mujhe lagta hai ki aaj ka din kaafi acha hai aur hum sabko kaam complete karna chahiye"

translated_text = translator.translate(hinglish_text, src='hi', dest='en')

print("Hinglish Text:", hinglish_text)
print("English Translation:", translated_text.text)
```

```python
english_text=translated_text.text
```

```python
doc=nlp(english_text)
displacy.render(doc,style='ent',jupyter=True)
```

```python
import re

def basic_clean(text):
    text = text.lower()
    text = re.sub(r'<.*?>', '', text)          # remove HTML tags
    text = re.sub(r'[^a-zA-Z\s]', '', text)    # keep only letters + space
    text = re.sub(r'\s+', ' ', text).strip()   # remove extra spaces
    return text

clean=basic_clean(text2)
```

```python
print(clean)
```

```python
doc=nlp(clean)

for ent in doc.ents:
    print(ent.text,"->",ent.label_)
```
